```python
# Ans : Box plots show the five-number summary of a set of data: including the minimum score, first (lower) quartile, 
#       median, third (upper) quartile, and maximum score.
#       The whiskers each cover 25% of the data values. The lower whisker covers all the data values from the minimum value up
#       to Q1, that is, the lowest 25% of data values. The upper whisker covers all the data values between Q3 and the maximum 
#       value, that is, the highest 25% of data values.
```
